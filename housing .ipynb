```python
varlist=["hotwaterheating","prefarea","airconditioning","guestroom","basement","mainroad"]
housing[varlist]=housing[varlist].apply(lambda x: x.map({'yes':1,'no':0}))
housing[varlist].head(10)
```

```python
#creating dummy variable for furninshing status
status=pd.get_dummies(housing['furnishingstatus'],drop_first=True)
status
```

```python
#concatinate status in housing
housing=pd.concat([housing,status], axis=1)
```

```python
housing.head()
```

```python
housing.drop('furnishingstatus', axis=1,inplace = True)
housing
```

## Variance Inflation Factor(VIF)

```python
#create a datafram thet woll contain the name of the feature as well as their respected vif's.
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif['Features']=x_train.columns
vif['VIF']=[variance_inflation_factor(x_train.values,i) for i in range(x_train.shape[1])]
vif['VIF']=round(vif['VIF'],2)
vif=vif.sort_values(by='VIF', ascending=False)
vif
```
